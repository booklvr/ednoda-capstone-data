````python
from pathlib import Path

import json
import pandas as pd
````

## 1. Check That The Processed Files Exist

This cell confirms that the core processed outputs are available.

````python
processed = Path('data/processed')
expected_files = [
    'sentences.parquet',
    'vocabulary_reference.parquet',
    'grammar_reference.parquet',
    'validation_summary.json',
    'build_manifest.json',
]

for name in expected_files:
    path = processed / name
    print(f'{name}:', 'FOUND' if path.exists() else 'MISSING')
````

## 2. Load The Main Tables

These are the main research tables students will work with:
- `sentences`
- `vocabulary_reference`
- `grammar_reference`

````python
sentences = pd.read_parquet(processed / 'sentences.parquet')
vocabulary = pd.read_parquet(processed / 'vocabulary_reference.parquet')
grammar = pd.read_parquet(processed / 'grammar_reference.parquet')

print('sentences:', len(sentences))
print('vocabulary_reference:', len(vocabulary))
print('grammar_reference:', len(grammar))
````

## 9. Optional: Load Sentence Features

If you already ran `python scripts/enrich_sentence_features.py`, you can load extra NLP features here.

This table includes:
- token counts
- lexical density
- POS-derived counts
- readability metrics

````python
sentence_features_path = processed / 'sentence_features.parquet'
sentence_features = pd.read_parquet(sentence_features_path) if sentence_features_path.exists() else None
print('sentence_features available:', sentence_features is not None)
````

````python
if sentence_features is not None:
    display(sentence_features[[
        'text_normalized',
        'cefr_level',
        'surface_token_count',
        'lexical_density',
        'noun_count',
        'verb_count',
        'readability_flesch_reading_ease',
        'root_token',
        'root_pos',
    ]].head(15))
else:
    print('Run: python scripts/enrich_sentence_features.py')
````

## 10. Simple Feature Analysis

This is an example of a very normal ML workflow question:

> Do some features vary by CEFR level?

Below we compare average lexical density by CEFR level.

````python
if sentence_features is not None:
    sentence_features.groupby('cefr_level')['lexical_density'].mean().sort_index()
else:
    print('Run: python scripts/enrich_sentence_features.py')
````
