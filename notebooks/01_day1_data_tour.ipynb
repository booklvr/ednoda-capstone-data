## 1. Setup

Run this cell first. It imports the libraries and figures out where the repo root is.

````python
from pathlib import Path
import json
import pandas as pd

cwd = Path.cwd()
project_root = cwd if (cwd / 'data').exists() else cwd.parent
processed = project_root / 'data' / 'processed'

print('Notebook cwd:', cwd)
print('Project root:', project_root)
print('Processed dir:', processed)
````

## 2. Check That The Processed Files Exist

This confirms that the pipeline outputs are present.

````python
expected_files = [
    'sentences.parquet',
    'vocabulary_reference.parquet',
    'grammar_reference.parquet',
    'validation_summary.json',
    'build_manifest.json',
]

for name in expected_files:
    path = processed / name
    print(f'{name}:', 'FOUND' if path.exists() else 'MISSING')
````

## 3. Load The Main Tables

These are the main processed tables students will work with.

````python
sentences = pd.read_parquet(processed / 'sentences.parquet')
vocabulary = pd.read_parquet(processed / 'vocabulary_reference.parquet')
grammar = pd.read_parquet(processed / 'grammar_reference.parquet')

print('sentences:', len(sentences))
print('vocabulary_reference:', len(vocabulary))
print('grammar_reference:', len(grammar))
````

## 10. Optional: Load Sentence Features

If you already ran `python scripts/enrich_sentence_features.py`, you can load the NLP feature table.

````python
sentence_features_path = processed / 'sentence_features.parquet'
print('sentence_features exists:', sentence_features_path.exists())

sentence_features = pd.read_parquet(sentence_features_path) if sentence_features_path.exists() else None
````

````python
if sentence_features is not None:
    sentence_features[[
        'text_normalized',
        'cefr_level',
        'surface_token_count',
        'lexical_density',
        'noun_count',
        'verb_count',
        'readability_flesch_reading_ease',
        'root_token',
        'root_pos',
    ]].head(15)
else:
    print('Run this in the terminal first: python scripts/enrich_sentence_features.py')
````

## 11. Simple Feature Analysis

A very normal ML question is: do some features vary by CEFR level?

Below we compare average lexical density by CEFR level.

````python
if sentence_features is not None:
    sentence_features.groupby('cefr_level')['lexical_density'].mean().sort_index()
else:
    print('Run this in the terminal first: python scripts/enrich_sentence_features.py')
````
